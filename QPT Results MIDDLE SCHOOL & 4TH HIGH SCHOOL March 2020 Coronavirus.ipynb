```python
def Calificar_Examenes_Clave(resultado_por_fila):
    
    filas_de_datos = resultado_por_fila[2:42]
    counter=0
    score=0
    if resultado_por_fila[1] == "A":
        counter=0
        score=0
        for fila in filas_de_datos:
            if claveA[counter] == filas_de_datos[counter]:
                score += 1
            counter += 1
    #return score
    elif resultado_por_fila[1] == "B":
        counter=0
        score=0
        for fila in filas_de_datos:
            if claveB[counter] == filas_de_datos[counter]:
                score += 1
            counter += 1
    return score
```

```python
test_results["SCORE"] = test_results.apply(Calificar_Examenes_Clave, axis="columns")
```

```python
test_results
```

```python
def Label_Level(score):
    fila= score[-1]
    if fila <= 8:
        return "Pre-A1"
    elif fila > 8 and fila <= 15:
        return "A1"
    elif fila > 15 and fila <= 23:
        return "A2"
    elif fila > 23 and fila <= 32:
        return "B1"
    elif fila > 32 and fila <= 40:
        return "B2"
```

```python
def Write_Grade(score):
    fila= score[-2]
    #if fila <= 8:
    #    return "Pre-A1"
    if fila <= 15:
        return "8"
    elif fila > 15 and fila <= 23:
        return "9"
    elif fila > 23:
        return "10"
    #elif fila > 32 and fila <= 40:
    #   return "B2"
```

```python
def Write_Grado_(score):
    fila= score[-4]
    return fila[0]
    
```

```python
test_results["NIVEL2"] = test_results.apply(Label_Level, axis="columns")
test_results["Grado"] = test_results.apply(Write_Grado_, axis="columns")
'test_results["CALIFICACION EXAMEN DE COLOCACION"] = test_results.apply(Write_Grade, axis="columns")'
test_results=test_results.sort_values("SCORE", axis=0, ascending= False)
test_results
```

```python
test_results_for_teachers=test_results.loc[:, ['FOLIO','VERSION','NOMBRE','PROFESOR','GRUPO','Grado','NIVEL','SCORE','NIVEL2']]
test_results_for_teachers
```
